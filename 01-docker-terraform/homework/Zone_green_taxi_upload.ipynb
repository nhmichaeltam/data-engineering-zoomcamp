```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
url_zones = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"
url_taxi2025 = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
```

```python
df_zones = pd.read_csv(url_zones)
df_taxi2025 = pd.read_parquet(url_taxi2025)
```

```python
df_zones.to_sql(con=engine, name="zones", if_exists="replace")
df_taxi2025.to_sql(con=engine, name="green_taxi_trips_2025_11", if_exists="replace")
```
